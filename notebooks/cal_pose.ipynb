```python

from pathlib import Path
from Data_storm_2025.src.video_analyzer import VideoAnalyzer
input_video_name = "Untitled00014096.mp4"

Path("results").mkdir(exist_ok=True)
input_video_path = Path(input_video_name)
output_video_path = Path(f"results/simple_analyzed_{input_video_name}")

if not input_video_path.exists():
    print(f"Không tìm thấy file video '{input_video_name}'!")
    print(f"Kéo thả file vào đúng thư mục 'Data_storm_2025'.")
else:
    analyzer = VideoAnalyzer()
    analyzer.generate_analysis_video(
        video_path=input_video_path,
        output_path=output_video_path,
        slowdown_factor=1.4
    )
```

```python
from IPython.display import HTML
from base64 import b64encode
import os

source_path = str(output_video_path)
web_compatible_path = "results/video_for_web.mp4"

if os.path.exists(source_path):
    print("⏳ Đang xử lý video để hiển thị...")

    os.system(f"ffmpeg -i {source_path} -vcodec libx264 {web_compatible_path} -y -loglevel error")

    mp4 = open(web_compatible_path, 'rb').read()
    data_url = "data:video/mp4;base64," + b64encode(mp4).decode()

    display(HTML(f"""
    <h3>Kết quả:</h3>
    <video width=640 controls autoplay>
          <source src="{data_url}" type="video/mp4">
    </video>
    """))
else:
    print(f" Không tìm thấy file: {source_path}")
```
